```python
df.groupby('Gender')['Churned'].mean()
```

```python
df.groupby('Membership_Years')['Churned'].mean().sort_values(ascending=False)
```

```python
df.groupby('Login_Frequency')['Churned'].mean().sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt

df['Churned'].value_counts().plot(kind='bar')
plt.title("Customer Churn Distribution")
plt.xlabel("Churned (0 = Stay, 1 = Leave)")
plt.ylabel("Number of Customers")
plt.show()
```

```python
import matplotlib.pyplot as plt

df.groupby('Customer_Service_Calls')['Churned'].mean().plot(kind='line')
plt.title("Churn Rate vs Customer Service Calls")
plt.xlabel("Customer Service Calls")
plt.ylabel("Churn Rate")
plt.show()
```
